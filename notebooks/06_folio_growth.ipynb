```python
# Folio count growth — line chart from 13.26 Cr (Jan 2022) to 26.12 Cr (Dec 2025). Mark key milestones.
```

```python
import pandas as pd
import plotly.express as px

folio_df = pd.read_csv(
    "../data/raw/06_industry_folio_count.csv",
    parse_dates=["month"]
)
```

```python
folio_df.head()
```

```python
folio_df.info()
```

```python
folio_df['month_name'] = folio_df['month'].dt.month_name()
folio_df['year'] = folio_df['month'].dt.year
```

```python
folio_df.head()
```

```python
fig = px.line(
    folio_df,
    x='month',
    y='total_folios_crore',
    title='Folio Count Growth (Jan 2022 - Dec 2025)',
    labels={'month': 'Month', 'total_folios_crore': 'Folio Count (in Crores)'},
)
fig.show()
```

```python
fig.add_annotation(
    x = folio_df.iloc[0]['month'],
    y = folio_df.iloc[0]['total_folios_crore'],
    text = "Jan 2022: 13.26 Cr",
    showarrow = True,
)

fig.add_annotation(
    x = folio_df.iloc[-1]['month'],
    y = folio_df.iloc[-1]['total_folios_crore'],
    text = "Dec 2025: 26.12 Cr",
    showarrow = True,
)
```
