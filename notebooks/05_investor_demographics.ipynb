```python
# investor demographics - age group distribution pie chart SIP amount boxplot by age group . gender split
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

investor_df = pd.read_csv("../data/processed/08_investor_transactions_clean.csv",
                          parse_dates=['transaction_date'])
investor_df.head()
```

```python
investor_df['age_group'].value_counts()
```

```python
age_counts = investor_df['age_group'].value_counts()

fig = px.pie(
    values= age_counts.values,
    names = age_counts.index,
    title='Investor distribution by age group'
)

fig.show()
```

```python
# SIP AMOUNT BOX PLOT BY AGE GROUP - how much does each group invest

sip_Transactions = investor_df[investor_df['transaction_type'] == 'SIP']
```

```python
fig = px.box(
    sip_Transactions,
    x='age_group',
    y='amount_inr',
    title = 'SIP amount distribution by age group',

    labels={
        'amount_inr':"Amount",
        'age_group': 'Age Group'
    }

)
fig.show()
```

```python
# gender split

investor_df['gender'].value_counts()
```

```python
gender_counts = investor_df['gender'].value_counts()

fig = px.pie(
    values= gender_counts.values,
    names= gender_counts.index,
    title='Investor distribution by gender'
)
fig.show()
```

```python
#Horizontal Bar Chart → SIP Amount by State
```

```python
investor_df.head()
```

```python
investor_df['state'].value_counts()
```

```python
sip_amount_by_state = (
    investor_df[investor_df['transaction_type'] == 'SIP']
    .groupby('state')['amount_inr']
    .sum()
    .sort_values(ascending=False)
)

sip_amount_by_state
```

```python
# bar chart of SIP amount by state
fig = px.bar(
    sip_amount_by_state,
    x=sip_amount_by_state.index,
    y=sip_amount_by_state.values,
    title='SIP Amount by State',
    labels={
        'x': 'State',
        'y': 'SIP Amount (INR)'
    }
)
fig.show()
```
