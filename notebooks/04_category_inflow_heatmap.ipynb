```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

category_df = pd.read_csv("../data/raw/05_category_inflows.csv",
                          parse_dates=['month'])
category_df.head()
```

```python
category_df['category'].value_counts()


category_df['month_name'] = category_df['month'].dt.month_name()
category_df.head()
```

```python
# creating the pivot table

heatmap_data = category_df.pivot(
    index='category',
    columns='month_name',
    values='net_inflow_crore'
)
heatmap_data.head()
```

```python
plt.figure(figsize=(12,6))

sns.heatmap(
    heatmap_data,
    cmap='YlGnBu', 
)
plt.title("Monthly net inflow by mutual fund category")
plt.xlabel("Month")
plt.ylabel("Category")

plt.show()
```
