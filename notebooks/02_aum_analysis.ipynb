```python
# AUM growth bar chart — grouped bar by fund house for each year (2022–2025). Highlight SBI at ₹12.5L Cr dominance using Seaborn.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
aum_df = pd.read_csv("../data/raw/03_aum_by_fund_house.csv",
                     parse_dates=['date']
)

aum_df.head()

aum_df.info()
```

```python
aum_df.head()
```

```python
aum_df["year"] = aum_df["date"].dt.year
aum_df['year'].unique()
```

```python
aum_df.head()
```

```python
plt.figure(figsize=(12,6))

sns.barplot(
    data=aum_df,
    x="year",
    y="aum_lakh_crore",
    hue="fund_house",
    palette = ['royalblue']+['lightgray']*9
   
)
plt.text(
    3,
    12.8,
    "SBI: rs 12.5 lakh cr",
    ha = 'center',
    fontsize = 11,
    fontweight = 'bold'
)
plt.title(
    "Assets Under Management by Fund House (2022–2025)",
    fontsize=16,
    fontweight="bold"
)
plt.xlabel("Year")
plt.ylabel("AUM (₹ Lakh Crore)")
plt.tight_layout()
plt.show()
```

### Business Insights

- SBI Mutual Fund maintained the highest Assets Under Management (AUM) throughout 2022–2025.
- SBI's AUM increased from approximately ₹6.05 lakh crore in 2022 to ₹12.5 lakh crore in 2025, reinforcing its market leadership.
- All major fund houses experienced consistent AUM growth, indicating strong expansion of the mutual fund industry.
- HDFC Mutual Fund and ICICI Prudential MF remained the closest competitors, but the gap with SBI widened over time.
