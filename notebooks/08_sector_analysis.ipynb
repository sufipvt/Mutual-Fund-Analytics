```python
# Across all mutual funds, which sectors receive the largest investment?"

import pandas as pd
import plotly.express as px

holdings_df = pd.read_csv("../data/raw/09_portfolio_holdings.csv")
```

```python
holdings_df.head()
```

```python
sector_weights = (holdings_df.groupby('sector')['weight_pct'].sum().reset_index().sort_values(by='weight_pct', ascending=False))
```

```python
sector_weights.head()
```

```python
fig = px.pie(
    sector_weights,
    values='weight_pct',
    names='sector',
    hole = 0.45,
    title = 'Sector Allocation Across All Mutual Funds'
)
fig.show()

fig.update_traces(textposition='inside', textinfo='percent+label')

fig.show()
```
