```python
import pandas as pd
import plotly.express as px

sip_df = pd.read_csv("../data/raw/04_monthly_sip_inflows.csv",
                     parse_dates=['month'])
sip_df.head()
```

```python
sip_df.info()
```

```python
fig = px.line(
    sip_df,
    x='month',
    y='sip_inflow_crore',
    title='Monthly SIP inflow trend (2022-2025)',
    labels={
        "month":"Month",
        "sip_inflow_crore":"SIP Inflow (Crore)"
    }
)
fig.show()
```

```python
max_row = sip_df.loc[sip_df['sip_inflow_crore'].idxmax()]
max_row
```

```python
sip_df.loc[47]
```

```python
# add the annotation
fig.add_annotation(
    x=max_row['month'],
    y = max_row['sip_inflow_crore'],
    text = 'All time high <br> Rs 31,002 Cr',
    showarrow=True,
    arrowhead=2
)
```
